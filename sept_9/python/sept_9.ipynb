```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# https://www.ncdc.noaa.gov/cag/time-series/global/globe/land_ocean/ytd/12/1880-2020.csv
# Global Land and Ocean Temperature Anomalies, January-December
# Units: Degrees Celsius
# Base Period: 1901-200
T = pd.read_csv('data/t.csv',names = ['year','value'])
# pd.read_csv?
# From https://data.giss.nasa.gov/modelforce/Fe_net_H11.txt
# Global Mean Net Forcing Relative to 1880
# Year      Net
F = pd.read_csv('data/f.csv',names = ['year','net'])
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True)
ax1.plot(T.year,T.value,'c')
ax1.set_ylabel('celcius')
ax1.set_title('Global Land and Ocean Temperature Anomalies')
ax2.plot(F.year,F.net,'m')
ax2.set_ylabel('net forcing')
ax2.set_title('Global Mean Net Forcing Relative to 1880')
ax2.set_xlabel('Years')
plt.show()
```

```python
T_mean = np.mean(T.value)
T_detrend = T.value - T_mean
F_mean = np.mean(F.net)
F_detrend = F.net - F_mean
length = len(F_detrend)
T_detrend = T_detrend[:length]


plt.plot(F_detrend,T_detrend, 'o')
m, b = np.polyfit(F_detrend,T_detrend, 1)
plt.plot(F_detrend, m*F_detrend + b)
plt.ylabel('celcius')
plt.xlabel('net forcing')
plt.title('Detrended Temperature vs. Detrended Net Forcing')
plt.show()
```

Climate sensitivity is the average change in the Earth's surface temperature in response to changes in radiative forcing (wikipedia). The linear regression performed in (c) suggests that as mean net forcing (relative to 1880) increases by one unit, then there is a .3 increase in global average temperature (relative to 1901). In this sense, this regression is a good indicator because it represents average change. However, because climate scientists’ uncertainty about the range of climate sensitivity is significant, this model is not a good indicator. Climate sensitivity could be high or low, and the climate’s response to an increase in net forcing could be non-linear.
